```python
# Here we are importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Here we are making the copy of dataframe
df1=df.copy()
```

```python
# Checking missing values
df1.isnull().sum()
```

```python
# Company column has 94 % and agent has 13 % missing values, its better to drop it rather than treating them
df1.drop(['agent','company'],axis=1,inplace=True)
```

```python
df1['children'].value_counts()
```

```python
df1['children'].mode()
```

```python
# Here we are filling the country with not known and children with mode that is 0
nan_replacements={'country':"not known","children":0}
df1=df1.fillna(nan_replacements)
```

```python
# what is the percentage of lead time for arrival date month
top_10_arrival = df1.groupby('arrival_date_month')['lead_time'].size().reset_index().sort_values(by="lead_time",ascending=False)
top_10_arrival
```

```python
total_amount= top_10_arrival['lead_time'].sum()
total_amount
```

```python
percentage_of_top_10_arrival = (top_10_arrival['lead_time']/total_amount)*100
percentage_of_top_10_arrival.reset_index().sort_values(by="lead_time",ascending=False).head(10)
```

```python
# Print the percentage with 2 decimal place after roundingoff
percentage_of_top_10_arrival = ((top_10_arrival['lead_time']/total_amount)*100).round(2)
percentage_of_top_10_arrival.reset_index().sort_values(by="lead_time",ascending=False).head(10)
```

```python
# Lets combine children and babies together as kids
df1['kids'] = (df1.children) + (df1.babies)
df1['kids'].reset_index()
```

```python
# Combine total members by adding kids and adults
df1['total_members'] = (df1.kids) + (df1.adults)
df1['total_members'].reset_index()
```

```python
# What is the waiting period for each customer who have done deposit

country_visitors = df1.groupby('country')['is_canceled'].size().reset_index().sort_values(by="is_canceled",ascending=False)
country_visitors
```

```python
# What is the waiting period for each customer who have done deposit and make a plot on it.

waiting_list = df1.groupby('deposit_type')['days_in_waiting_list'].size().reset_index().sort_values(by='days_in_waiting_list',ascending=False).plot()
waiting_list
```

```python
 bookings_confirmed = df1[df1['is_canceled'] == 0]
bookings_confirmed
```

```python
#most of the visitors prefer bed & breakfast (BB) followed by full board (FB), half board (HB) and self cater (SC)
plt.figure(figsize=(15,6))
df_pie = bookings_confirmed.groupby(['adults']).agg({'hotel':'count'}).reset_index()
colors = sns.color_palette('bright')[0:5]
plt.figure(figsize=(10,4))

plt.pie(df_pie['hotel'], labels = df_pie.adults ,colors=colors,autopct='%.0f%%')
plt.title('Percentage of adults booking')
plt.show()
```

```python
# Find out how many repeated guests are there in the hotel
type_of_guests = df1.groupby('is_repeated_guest')['hotel'].size().reset_index().sort_values(by="hotel",ascending=True)
type_of_guests
```
